```python
# TimHortons_Data.ipynb

# --- Step 1: Import Libraries ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Quick display settings
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")

# --- Step 2: Define Sales & Revenue Data ---
# Example quarterly sales figures (in millions USD)
sales_data = {
    "Quarter": ["Q1 2025", "Q2 2025", "Q3 2025", "Q4 2025"],
    "Revenue_USD_m": [1125, 1140, 1180, 1200],  # Example numbers
    "SameStoreSales_Growth_pct": [3.0, 4.5, 5.0, 4.0]
}
sales_df = pd.DataFrame(sales_data)
sales_df

# Plot Sales Trend
plt.figure(figsize=(8,5))
sns.lineplot(x="Quarter", y="Revenue_USD_m", data=sales_df, marker="o")
plt.title("Tim Hortons - Estimated Coffee & Beverage Revenue 2025")
plt.ylabel("Revenue (Million USD)")
plt.xlabel("Quarter")
plt.show()


# --- Step 3: Example Profit Margin Calculation ---
# Assume hypothetical gross margin and net profit margin
profit_data = {
    "Quarter": ["Q1 2025", "Q2 2025", "Q3 2025", "Q4 2025"],
    "GrossMargin_pct": [55, 54, 53, 55],
    "NetProfitMargin_pct": [10, 12, 11, 13]
}
profit_df = pd.DataFrame(profit_data)
profit_df

# Merge with sales
merged_df = pd.merge(sales_df, profit_df, on="Quarter")
merged_df

# Plot Margins
plt.figure(figsize=(8,5))
sns.barplot(x="Quarter", y="NetProfitMargin_pct", data=merged_df)
plt.title("Approximate Tim Hortons Net Profit Margins (Estimates)")
plt.ylabel("Net Profit Margin (%)")
plt.xlabel("Quarter")
plt.show()


# --- Step 4: Sample Demographics Data ---
# Example demographics of coffee drinkers / customers (survey-based)
demographics_data = {
    "AgeGroup": ["18-24", "25-34", "35-44", "45-54", "55+"],
    "Customer_pct": [15, 35, 25, 15, 10]
}
demo_df = pd.DataFrame(demographics_data)
demo_df

# Plot Demographics
plt.figure(figsize=(7,5))
sns.barplot(x="AgeGroup", y="Customer_pct", data=demo_df)
plt.title("Estimated Tim Hortons Coffee Customer Age Demographics")
plt.ylabel("Percent of Customers (%)")
plt.xlabel("Age Group")
plt.show()


# --- Step 5: Save Data for Future Use ---
sales_df.to_csv("sales_data.csv", index=False)
profit_df.to_csv("profit_data.csv", index=False)
demo_df.to_csv("demographics_data.csv", index=False)
print("Datasets saved locally.")
```
