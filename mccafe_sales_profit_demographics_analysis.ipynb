```python
# 1. Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# McDonald's revenue (sample years, USD billions)
mcd_revenue = {
    "Year":[2013, 2018, 2023, 2024, 2025],
    "McD_Revenue_Billions":[28.106, 21.258, 25.494, 25.92, 26.885]
}

df_revenue = pd.DataFrame(mcd_revenue)

# Estimated McCafé annual sales in US (in billions)
# Source indicates coffee sold through McCafé outlets generates ~$4B annually in the US.
mccafe_sales = {
    "Region":["US (estimate)","Worldwide (conceptual)"],
    "Sales_Billions":[4.0, None]   # worldwide not published, so placeholder
}
df_mccafe = pd.DataFrame(mccafe_sales)

df_revenue, df_mccafe
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=df_revenue, x="Year", y="McD_Revenue_Billions", marker="o")
plt.title("McDonald's Annual Revenue (Selected Years)")
plt.ylabel("Revenue (USD Billions)")
plt.xlabel("Year")
plt.grid(True)
plt.show()
```

```python
# Estimated McDonald’s operating margin
operating_margin = 46.9  # percent
net_margin    = 31.85   # percent from finance analysis

profit_df = pd.DataFrame({
    "Margin_Type":["Operating Margin %", "Net Margin %"],
    "Value":[operating_margin, net_margin]
})

profit_df
```

```python
plt.figure(figsize=(6,4))
sns.barplot(data=profit_df, x="Margin_Type", y="Value")
plt.title("McDonald's Estimated Profit Margins")
plt.ylabel("Percentage (%)")
plt.xlabel("Margin Type")
plt.show()
```

```python
demo_data = {
    "Age_Group": ["18–24", "25–34", "35–44", "45–54", "55+"],
    "Percent_of_Customers":[10, 30, 25, 20, 15],
    "Male_Perc":[48, 50, 52, 51, 49],
    "Female_Perc":[52, 50, 48, 49, 51]
}

df_demo = pd.DataFrame(demo_data)
df_demo
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df_demo, x="Age_Group", y="Percent_of_Customers", palette="viridis")
plt.title("Hypothetical McCafé Customer Age Distribution")
plt.ylabel("Percent of Customers (%)")
plt.xlabel("Age Group")
plt.show()
```

```python
df_gender = df_demo.melt(id_vars="Age_Group", value_vars=["Male_Perc","Female_Perc"], var_name="Gender", value_name="Percentage")

plt.figure(figsize=(8,5))
sns.barplot(data=df_gender, x="Age_Group", y="Percentage", hue="Gender")
plt.title("Hypothetical Gender Breakdown by Age Group")
plt.ylabel("Percentage (%)")
plt.xlabel("Age Group")
plt.show()
```

```python
# Save all datasets to CSV
df_revenue.to_csv("mcdonalds_revenue.csv", index=False)
df_mccafe.to_csv("mccafe_sales_estimates.csv", index=False)
profit_df.to_csv("mcd_profit_margins.csv", index=False)
df_demo.to_csv("mccafe_customer_demographics.csv", index=False)

print("Datasets saved.")
```
