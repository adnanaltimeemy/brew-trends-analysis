```python
# keurig_dr_pepper_analysis.ipynb

# ===============================
#  Step 1  Import Libraries
# ===============================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# ===============================
#  Step 2  Create a Financial Dataset
# ===============================
# These numbers are from recent company reports and aggregated sources:
#  Annual revenue (20222025)
#  Quarterly sales (2025)
#  Net profit margin

data_financial = {
    "Year": [2022, 2023, 2024, "2025 (est)"],
    "Annual_Revenue_USD_Billion": [14.057, 14.814, 15.351, 16.174],
    "Net_Profit_Margin_pct": [9.5, 9.8, 9.8, 9.8]  # Approx yearly margin from average estimates
}

financial_df = pd.DataFrame(data_financial)
display(financial_df)

# ===============================
#  Step 3  Add Quarterly Sales (2025)
# ===============================
# Quarterly sales data sourced from financial releases and trading data
data_quarters_2025 = {
    "Quarter": ["Q1", "Q2", "Q3", "Q4"],
    "Sales_Revenue_USD_Billion": [4.16, 2.7, 4.31, 4.50]
}

quarter_df = pd.DataFrame(data_quarters_2025)
display(quarter_df)

# ===============================
#  Step 4  Simulate Demographic Dataset
# ===============================
# Note: Public company data does not include direct "demographics" of consumers.
# This is a simulated example dataset showing demographic segments of consumers.

demo_data = {
    "Segment": ["18-24", "25-34", "35-44", "45-54", "55+"],
    "Percent_Sales": [10, 22, 25, 20, 23],
    "Avg_Annual_Spend_USD": [120, 250, 300, 280, 260]
}

demo_df = pd.DataFrame(demo_data)
display(demo_df)

# ===============================
#  Step 5  Visualize the Financial Trends
# ===============================

plt.figure(figsize=(10,5))
sns.barplot(x=financial_df["Year"].astype(str),
            y=financial_df["Annual_Revenue_USD_Billion"],
            palette="viridis")
plt.title("Keurig Dr Pepper Annual Revenue (20222025 Est.)")
plt.ylabel("Revenue (USD Billion)")
plt.xlabel("Fiscal Year")
plt.show()

# ===============================
#  Step 6  Plot Quarterly Sales
# ===============================

plt.figure(figsize=(8,4))
sns.lineplot(x="Quarter", y="Sales_Revenue_USD_Billion", 
             data=quarter_df, marker="o")
plt.title("KDP 2025 Quarterly Sales Revenue")
plt.ylabel("Revenue (USD Billion)")
plt.xlabel("Quarter")
plt.show()

# ===============================
#  Step 7  Demographic Spend Insight
# ===============================

plt.figure(figsize=(8,4))
sns.barplot(x="Segment", y="Avg_Annual_Spend_USD", data=demo_df)
plt.title("Simulated Consumer Demographics: Avg Spend by Age Segment")
plt.ylabel("Avg Annual Spend (USD)")
plt.xlabel("Age Group")
plt.show()
```
