```python
# ===========================
# ðŸ“Œ Notebook: GMCR Sales & Financial Dataset
# ===========================

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1ï¸âƒ£ Create high-level financial data
financial_data = {
    'Fiscal Year': [2010, 2011, 2012],
    'Net Sales (USD millions)': [1356.8, 2650.9, 3859.2],  # from SEC data excerpt for 2011â€“12
    'Gross Profit (%)': [np.nan, 34.1, 32.9],              # example margins from 2012 report :contentReference[oaicite:1]{index=1}
    'Operating Margin (%)': [np.nan, 13.9, 14.7],          # example figure from same source :contentReference[oaicite:2]{index=2}
}

df_financials = pd.DataFrame(financial_data)
print("ðŸ“Š Financial Data")
df_financials

# 2ï¸âƒ£ Add profit margin summary from external analysis
profit_margins = {
    'Year Range': ['~2012'],
    'Avg Gross Margin (%)': [34.1],   # from financial source summary :contentReference[oaicite:3]{index=3}
    'Avg Net Margin (%)': [np.nan]
}
df_margins = pd.DataFrame(profit_margins)
print("ðŸ“ˆ Profit Margin Summary")
df_margins

# 3ï¸âƒ£ Create demographic data (proxy example)
# Note: we do not have exact customer demographic data for Green Mountain sales, so we use publicly referenced market demographics
demo_data = {
    'Household Income Group': ['>= $70K', '< $70K'],
    'Total U.S. Households (%)': [63, 37],
    'Share of Single Cup Coffee Sales (%)': [55, 45]
}
df_demo = pd.DataFrame(demo_data)
print("ðŸ‘¥ Customer Demographic Proxy Data")
df_demo

# 4ï¸âƒ£ Combine into a single dataset structure
df_combined = df_financials.copy()
# Example columns to append demographic proxy data
df_combined['High Income Sales Share (%)'] = [65, 56, 55]  # hypothetical correlated values
df_combined['Low Income Sales Share (%)'] = [35, 44, 45]
print("ðŸ“š Combined Dataset")
df_combined

# 5ï¸âƒ£ Data visualization examples
plt.figure(figsize=(8, 5))
plt.plot(df_financials['Fiscal Year'], df_financials['Net Sales (USD millions)'], marker='o')
plt.title("GMCR Net Sales (2010â€“2012)")
plt.xlabel("Fiscal Year")
plt.ylabel("Net Sales (USD millions)")
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 5))
plt.bar(df_demo['Household Income Group'],
        df_demo['Share of Single Cup Coffee Sales (%)'],
        color=['green', 'orange'])
plt.title("Demographic Sales Share by Income Group")
plt.ylabel("Sales Share (%)")
plt.show()
```
