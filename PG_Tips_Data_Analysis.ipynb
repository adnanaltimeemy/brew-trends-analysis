```python
import pandas as pd
import numpy as np

# Create a mock dataset of annual sales and profit margins for PG Tips
data = {
    "Year": np.arange(2018, 2025),
    "Sales_GBP_million": [115, 112, 108, 104, 98, 94, 88],  # mock sales values
    "Profit_GBP_million": [12, 11, 10, 8, 6, 6.2, 7],        # mock profits
    "Market_Share_%": [30, 29, 28, 26, 24, 23, 22],         # trend of market share decline
    "Black_Tea_Consumption_%": [75, 74, 73, 70, 66, 64, 60],# estimates
}

df_sales = pd.DataFrame(data)
df_sales["Profit_Margin_%"] = (df_sales["Profit_GBP_million"] / df_sales["Sales_GBP_million"]) * 100

df_sales.head()
```

```python
# Demographic preferences by age segment (mock example)
demographics = {
    "Age_Group": ["18-24", "25-34", "35-44", "45-54", "55-64", "65+"],
    "Tea_Preference_%": [25, 32, 45, 58, 72, 80],
    "Coffee_Preference_%": [55, 47, 38, 30, 22, 15]
}

df_demo = pd.DataFrame(demographics)
df_demo.head()
```

```python
import matplotlib.pyplot as plt

# Plot Sales & Profit Trends
plt.figure(figsize=(10,6))
plt.plot(df_sales["Year"], df_sales["Sales_GBP_million"], marker='o', label="Sales (GBP m)")
plt.plot(df_sales["Year"], df_sales["Profit_GBP_million"], marker='o', label="Profit (GBP m)")
plt.title("PG Tips Sales & Profit (20182024)")
plt.xlabel("Year")
plt.ylabel("Value (GBP million)")
plt.legend()
plt.grid(True)
plt.show()

# Plot Demographic Tea Preference
plt.figure(figsize=(8,5))
plt.bar(df_demo["Age_Group"], df_demo["Tea_Preference_%"], color='green')
plt.title("Tea Preference by Age Group (Mock Data)")
plt.ylabel("Tea Preference %")
plt.xlabel("Age Group")
plt.show()
```
